# **INDEXING**

•	Indexing is the way to access individual elements of a sequence (like a string, list, or tuple) using their position.


•	Python sequences are ordered, so each element has a unique index.

image.png

```python
# String Indexing

s = "hello Python"

print(s[0])    # h
print(s[1])    # e
print(s[2])    # l
print(s[3])    # l
print(s[4])    # o

print(s[-1])   # d
print(s[-2])   # l
print(s[-3])   # r
```

```python
print(s[-11])
```

# **SLICING**

• Slicing allows you to extract a part of a sequence (like a string, list, or tuple).

• It is done using indexes in the format

```python
# String Slicing
#Slicing[start:stop:step]
s = "hello world"

print(s[::2])
```

```python
#check the entered text is palindrome

text = input("enter the text:")

if text == text[::-1]:
  print("Its palindrome")
```

```python
# Email Generator

# Taking input from user
full_name = input("Enter Full Name: ")
domain_name = input("Enter Domain Name: ")

# String operations
lower_name = full_name.lower()
email_name = lower_name.replace(" ", ".")

# Email creation
email = email_name + "@" + domain_name

# Display output
print("\n----- User Details -----")

print(f"Full Name       : {full_name.upper()}")
print(f"Generated Email : {email}")
```
